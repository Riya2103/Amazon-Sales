```python
df['Category']=df['Category'].astype(str)
c_d=df['Category']
plt.figure(figsize=(10,6))
plt.hist(c_d,bins=15,edgecolor="Black", color="Yellow")
plt.xticks(rotation=90)
plt.show()
```

```python
# checking b2b data
```

```python
check_b2b=df['B2B'].value_counts()
```

```python
# create pie chart
```

```python
plt.pie(check_b2b, labels=check_b2b.index, autopct='%1.1f%%')
plt.show()
```

99.2% of buyers are retailers & 0.8% of buyers are wholesalers.

```python
# scatter plot
```

```python
x_data=df['Category']
y_data=df['Size']

plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Available size')
plt.show()
```

```python
# distribution by state
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data=df, x='ship-state')
plt.xlabel('State')
plt.ylabel('Order count')
plt.title('Distribution by State')
plt.show()
```

```python
# top 10 states
```

```python
top10_state=df['ship-state'].value_counts().head(10)
plt.figure(figsize=(18,8))
sns.countplot(data=df[df['ship-state'].isin(top10_state.index)], x='ship-state', palette="Set2")
plt.xlabel('State')
plt.ylabel('Order count')
plt.title('Distribution by State')
plt.show()
```

Most of the buyers are from Maharashtra.
